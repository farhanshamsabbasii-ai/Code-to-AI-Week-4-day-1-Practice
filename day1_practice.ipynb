## Exercise 3 — Indexing & slicing
Given the array below, print:
- the first element
- the last element
- elements from position 2 to 5
- every element from position 3 to the end

```python
import numpy as np

nums = np.array([5, 10, 15, 20, 25, 30, 35, 40])

# YOUR CODE HERE
print(nums[0])
print(nums[-1])
print(nums[2:5])
print(nums[3:])
```
